# model validation

1. colinearity

```python
rsq_RDS = smf.ols('RDS~ADM+MKS',data = df1).fit().rsquared
vif_RDS =1/ (1-rsq_RDS)
rsq_ADM = smf.ols('ADM~RDS+MKS',data = df1).fit().rsquared
vif_ADM =1/ (1-rsq_ADM)
rsq_MKS= smf.ols('MKS~RDS+ADM',data = df1).fit().rsquared
vif_MKS =1/ (1-rsq_MKS)
```

```python
vif_RDS
```

```python
vif_ADM
```

```python
vif_MKS
```

since no vif id greater than 20 we can conclude there is no  colinearity problem.we can consider all columns

# Residual analysis

# Q-Q PLOT test for normality 

```python
import statsmodels.api as sm
qqplot = sm.qqplot(model.resid,line = 'q')
plt.title('normal Q-Q plot Residuals')
plt.show()
```

```python
def get_standardized_values(vals):
    return(vals-vals.mean()/vals.std())
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
           get_standardized_values(model.resid))
plt.title('residues plot')
plt.xlabel('model.fittedvalues')
plt.ylabel('model.resid')
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
fig = sm.graphics.plot_regress_exog(model, 'RDS', fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
fig = sm.graphics.plot_regress_exog(model, 'ADM', fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
fig = sm.graphics.plot_regress_exog(model, 'MKS', fig=fig)
plt.show()
```
